# undersampling

```python
normal = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
# normal.shape
normal_sample = normal.sample(n=473)
```

```python
fraud.shape
```

```python
new_data=pd.concat([normal_sample,fraud],ignore_index=True)
```

```python
new_data.shape
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
x = new_data.drop('Class',axis=1)
y = new_data['Class']
print(x.shape)
print(y.shape)
```

## STORE FEATURE MATRIX IN X AND RESPONSE (TARGET) IN Y

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## logestic regression 

```python
from sklearn.linear_model import LogisticRegression
lc = LogisticRegression()
lc.fit(x_train,y_train)
```

```python
y_pred = lc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
print(x_test.shape)  # Should print (55133, <num_features>)
print(y_pred.shape)  # Should print (55133,)
```

```python
from sklearn.metrics import confusion_matrix,recall_score,precision_score,f1_score

recall_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

##  RANDOM FOREST classfier

## over sampling 

```python
x = data.drop('Class',axis=1)
y = data['Class']
print(x.shape)
print(y.shape)
```

```python
from imblearn.over_sampling import SMOTE ## this is used ti balance our data
```

```python
x_res,y_res = SMOTE().fit_resample(x,y)
```

```python
x_res.value_counts()
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_res,y_res,test_size=0.2,random_state=42)
```

## LOGISTIC REGRESSION

```python
y_pred1 = log.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
precision_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```
